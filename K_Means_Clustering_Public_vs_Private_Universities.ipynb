```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data.drop(['Private','Unnamed: 0'],axis=1))
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# in a normal data set, you would not have the actual labels to compare to the KMeans labels, 
# but since we do in this data set, lets see how our model did
```

```python
data['Private_Category'] = 0

# Because K means is unsupervised, I did not specify which way the label of private == yes should be. 
# In this case it is opposite what you might expect, and so yes == 0 and no == 1 contrary to common practice. 
data.loc[data.index[data['Private'] == 'Yes'],'Private_Category'] = 0
data.loc[data.index[data['Private'] == 'No'],'Private_Category'] = 1
```

```python
data
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(data['Private_Category'],kmeans.labels_))
print('\n')
print(classification_report(data['Private_Category'],kmeans.labels_))
```

```python
# This model was able to predict if a college was private or public with reasonable accuracy 
# given that this is a completely unsupervised model using only features. 
```
